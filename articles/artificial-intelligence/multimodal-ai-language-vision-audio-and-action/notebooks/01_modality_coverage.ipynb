# Modality Coverage Review

```python
import pandas as pd

records = pd.DataFrame({
    "case_id": ["A", "B", "C"],
    "has_text": [1, 1, 1],
    "has_vision": [1, 0, 1],
    "has_audio": [0, 1, 1],
    "has_sensor": [1, 0, 0]
})

records["modality_count"] = records[["has_text", "has_vision", "has_audio", "has_sensor"]].sum(axis=1)
records["coverage_score"] = (records["modality_count"] / 4).clip(upper=1)

records
```
