# Representation Learning Sketch

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(42)

data = pd.DataFrame({
    "feature_1": rng.normal(0, 1, 500),
    "feature_2": rng.normal(0, 1, 500),
    "feature_3": rng.normal(0, 1, 500)
})

data["latent_representation"] = (
    0.5 * data["feature_1"]
    - 0.2 * data["feature_2"]
    + 0.8 * data["feature_3"]
)

data.head()
```
