# Contrastive Learning

```python
import numpy as np
import pandas as pd

pairs = pd.DataFrame({
    "pair_id": ["P001", "P002", "P003"],
    "positive_similarity": [0.82, 0.76, 0.64],
    "negative_similarity": [0.25, 0.48, 0.59]
})

pairs["contrastive_margin"] = pairs["positive_similarity"] - pairs["negative_similarity"]
pairs
```
