# Slice-Level Monitoring

```python
import pandas as pd

slices = pd.DataFrame({
    "slice": ["overall", "region_a", "region_b", "new_source"],
    "accuracy": [0.84, 0.86, 0.79, 0.69],
    "missing_rate": [0.02, 0.01, 0.03, 0.11]
})

slices["review_required"] = (
    (slices["accuracy"] < 0.78) |
    (slices["missing_rate"] > 0.08)
)

slices
```
